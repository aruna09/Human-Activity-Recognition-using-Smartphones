```python
import pandas as pd
import numpy as np
```

```python
# dataset 1
raw_traindata = pd.read_csv("train.csv")
raw_testdata = pd.read_csv("test.csv")
for col in raw_testdata:
    print(col)
```

```python
# changing labels to have numeric values for ease of fitting the model

for row, index in raw_traindata.iterrows():
    if raw_traindata.iat[row, 562] == "WALKING":
        raw_traindata.iat[row, 562] = 0
    if raw_traindata.iat[row, 562] == "WALKING_UPSTAIRS":
        raw_traindata.iat[row, 562] = 1
    if raw_traindata.iat[row, 562] == "WALKING_DOWNSTAIRS":
        raw_traindata.iat[row, 562] = 2
    if raw_traindata.iat[row, 562] == "SITTING":
        raw_traindata.iat[row, 562] = 3
    if raw_traindata.iat[row, 562] == "STANDING":
        raw_traindata.iat[row, 562] = 4
    if raw_traindata.iat[row, 562] == "LAYING":
        raw_traindata.iat[row, 562] = 5

for row, index in raw_testdata.iterrows():
    if raw_testdata.iat[row, 562] == "WALKING":
        raw_testdata.iat[row, 562] = 0
    if raw_testdata.iat[row, 562] == "WALKING_UPSTAIRS":
        raw_testdata.iat[row, 562] = 1
    if raw_testdata.iat[row, 562] == "WALKING_DOWNSTAIRS":
        raw_testdata.iat[row, 562] = 2
    if raw_testdata.iat[row, 562] == "SITTING":
        raw_testdata.iat[row, 562] = 3
    if raw_testdata.iat[row, 562] == "STANDING":
        raw_testdata.iat[row, 562] = 4
    if raw_testdata.iat[row, 562] == "LAYING":
        raw_testdata.iat[row, 562] = 5
```

```python
train_features = raw_traindata.iloc[:, 0:562]
train_target = raw_traindata.iloc[:, 562]
train_target=train_target.astype('int')
test_features = raw_testdata.iloc[:, 0:562]
test_target = raw_testdata.iloc[:, 562]
test_target=test_target.astype('int')

#temp = np.array(test_features).reshape((1, -1))
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

clf = LogisticRegression(solver='lbfgs',multi_class='multinomial').fit(train_features, train_target)
predicted = clf.predict(test_features)
accuracy_score(test_target, predicted)
```

```python
from sklearn import tree
clf = tree.DecisionTreeClassifier().fit(train_features, train_target)
predicted = clf.predict(test_features)
accuracy_score(test_target, predicted)
```
